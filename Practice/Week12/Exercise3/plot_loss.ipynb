```python
import glob
import os
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

# Define the path pattern to find train_loss.csv files
files = glob.glob(os.path.join(os.path.expanduser('~'), 'CompChempractice', '*', 'train_loss.csv'))
test_files = glob.glob(os.path.join(os.path.expanduser('~'), 'CompChempractice', '*', 'test_loss.csv'))
# Plotting setup
plt.figure(figsize=(6, 6))
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('Iteration', fontsize=16)
plt.ylabel('Loss', fontsize=16)
print('train_loss')
# Iterate over the files
for file in files:
    name = file.split('/')[-2]
    df = pd.read_csv(file)
    x = df['Iteration']
    y = df['Loss']
    cubic_interpolation_model = interp1d(x, y, kind="cubic")
    Y_ = cubic_interpolation_model(x)
    plt.plot(x, Y_, label=name)
# Add legend
plt.legend(fontsize=12)
plt.show()
print('test_loss')

plt.figure(figsize=(6, 6))
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('Iteration', fontsize=16)
plt.ylabel('Loss', fontsize=16)
for file in test_files:
    name = file.split('/')[-2]
    df = pd.read_csv(file)
    x = df['Iteration']
    y = df['Loss']
    cubic_interpolation_model = interp1d(x, y, kind="cubic")
    Y_ = cubic_interpolation_model(x)
    plt.plot(x, Y_, label=name)
# Add legend
plt.legend(fontsize=12)
plt.show()
```
