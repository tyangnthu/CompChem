```python
from rdkit import Chem
from rdkit.Chem import PandasTools
import glob
import pandas as pd
from tqdm import tqdm

pd.set_option('display.max_rows', None)

files = glob.glob('*mol2')
smis = []
refcodes = []
for file in tqdm(files):
    refcode = file.split('.')[0]
    refcodes.append(refcode)
    mol = Chem.MolFromMol2File(file)
    smi = Chem.MolToSmiles(mol)
    smis.append(smi)
df = pd.DataFrame(list(zip(refcodes, smis)), columns=['refcodes', 'smiles'])
PandasTools.AddMoleculeColumnToFrame(df, 'smiles', 'molecules')
df
```
